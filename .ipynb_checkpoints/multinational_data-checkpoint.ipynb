```python

import psycopg2
import yaml
from sqlalchemy import create_engine
import pandas as pd

# Create a class that will contain the methods to connect to the database

class DatabaseConnector:
    def __init__(self, config_file):
        self.config_file = config_file

    def read_db_creds(self):
        with open(self.config_file, 'r') as file:
            creds = yaml.safe_load(file)
        return creds
    def init_db_engine(self):
        creds = self.read_db_creds()
        username = creds['RDS_USER']
        password = creds['RDS_PASSWORD']
        port = creds['RDS_PORT']
        host = creds['RDS_HOST']
        db_name = creds['RDS_DATABASE']
        conn_url = f"postgresql://{username}:{password}@{host}:{port}/{db_name}"
        engine = create_engine(conn_url)
        return engine
    def list_db_tables(self):
        query = """
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'public';
"""
        engine = self.init_db_engine()
        with engine.connect() as connection:
            table_names = connection.execute(query)
            return [table[0] for table in table_names]

initilize_class_connection = DatabaseConnector('db_creds.yaml')
initilize_read_creditial = initilize_class_connection.read_db_creds()
engine_create_initialization = initilize_class_connection.init_db_engine()
list_of_db_tables = initilize_class_connection.list_db_tables()
list_of_db_tables[1]
```

```python
import pandas as pd
from database_utils import DatabaseConnector

def read_table_to_dataframe(db_connector, table_name):
    # Check if the specified table exists in the database
    table_names = db_connector.list_db_tables()
    if table_name not in table_names:
        raise ValueError(f"Table '{table_name}' not found in the database.")

    # Initialize the database engine
    engine = db_connector.init_db_engine()

    # Read the table into a Pandas DataFrame
    query = f"SELECT * FROM {table_name}"
    df = pd.read_sql_query(query, engine)

    return df

# Create an instance of the DatabaseConnector
db = DatabaseConnector('db_creds.yaml')

# Specify the table name you want to read
table_name = 'legacy_users'

# Call the function to read the table and get a Pandas DataFrame
table_dataframe = read_table_to_dataframe(db, table_name)

# Now you can work with the table_dataframe as a Pandas DataFrame
table_dataframe.head(10)

```

```python
import pandas as pd
from database_utils import DatabaseConnector

class DataExtractor:
    def __init__(self, db_connector):
        self.db_connector = db_connector

    def read_rds_table(self, table_name):
        # Check if the specified table exists in the database
        table_names = self.db_connector.list_db_tables()
        if table_name not in table_names:
            raise ValueError(f"Table '{table_name}' not found in the database.")

        # Initialize the database engine
        engine = self.db_connector.init_db_engine()

        # Read the table into a Pandas DataFrame
        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql_query(query, engine)

        return df

# Create an instance of the DatabaseConnector
db = DatabaseConnector('db_creds.yaml')

# Create an instance of the DataExtractor class
data_extractor = DataExtractor(db)

# Specify the table name you want to read
table_name = 'legacy_store_details'

# Call the method to read the table and get a Pandas DataFrame
df = data_extractor.read_rds_table(table_name)

# Now you can work with the table_dataframe as a Pandas DataFrame
df
```
