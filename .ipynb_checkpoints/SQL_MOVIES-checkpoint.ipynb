```python
WEEK = '''
SELECT CASE
           WHEN EXTRACT(DOW FROM payment_date) = 0 THEN 'Sunday'
           WHEN EXTRACT(DOW FROM payment_date) = 1 THEN 'Monday'
           WHEN EXTRACT(DOW FROM payment_date) = 2 THEN 'Tuesday'
           WHEN EXTRACT(DOW FROM payment_date) = 3 THEN 'Wednesday'
           WHEN EXTRACT(DOW FROM payment_date) = 4 THEN 'Thursday'
           WHEN EXTRACT(DOW FROM payment_date) = 5 THEN 'Friday'
           WHEN EXTRACT(DOW FROM payment_date) = 6 THEN 'Saturday'
       END AS day_of_week,
       SUM(amount) AS amount_sumation
FROM payment
GROUP BY day_of_week
ORDER BY amount_sumation DESC;
'''
dr = pd.read_sql_query(WEEK, engine)
dr
```

```python
week = '''SELECT CASE
           WHEN EXTRACT(DOW FROM payment_date) = 0 THEN 'Sunday'
           WHEN EXTRACT(DOW FROM payment_date) = 1 THEN 'Monday'
           WHEN EXTRACT(DOW FROM payment_date) = 2 THEN 'Tuesday'
           WHEN EXTRACT(DOW FROM payment_date) = 3 THEN 'Wednesday'
           WHEN EXTRACT(DOW FROM payment_date) = 4 THEN 'Thursday'
           WHEN EXTRACT(DOW FROM payment_date) = 5 THEN 'Friday'
           WHEN EXTRACT(DOW FROM payment_date) = 6 THEN 'Saturday'
       END AS day_of_week,
       SUM(amount) AS amount_sumation, COUNT(customer_id) AS number_of_purchases
FROM payment
GROUP BY day_of_week
ORDER BY amount_sumation DESC;'''

goal = pd.read_sql_query(week, engine)
goal
```

```python
num = '''SELECT 
    s.store_id,
    a.address AS store_address,
    CONCAT(st.first_name, ' ', st.last_name) AS store_manager_name,
    SUM(p.amount) AS total_income
FROM store s
JOIN address a ON s.address_id = a.address_id
JOIN staff st ON s.manager_staff_id = st.staff_id
JOIN payment p ON st.staff_id = p.staff_id
GROUP BY s.store_id, a.address, store_manager_name;'''
df = pd.read_sql_query(num, engine)
df
```

```python
num = '''SELECT c.city AS city_name, SUM(p.amount) AS total_amount_spent
FROM city c
JOIN address a ON a.city_id = c.city_id
JOIN customer cm ON cm.address_id = a.address_id
JOIN payment p ON p.customer_id = cm.customer_id
JOIN rental r ON r.customer_id = cm.customer_id
WHERE p.payment_date >= r.rental_date
GROUP BY c.city
HAVING SUM(p.amount) > 150
ORDER BY c.city;'''

df = pd.read_sql_query(num, engine)
df.head(10)
```

```python
num = '''SELECT
    CASE
        WHEN p.amount < 3 THEN 'Low'
        WHEN p.amount >= 3 AND p.amount <= 7 THEN 'Medium'
        ELSE 'High'
    END AS transaction_category,
    COUNT(DISTINCT p.customer_id) AS customer_count
FROM payment p
GROUP BY transaction_category;'''
df = pd.read_sql_query(num, engine)
df
```

```python
num = '''SELECT f.film_id, f.title, COUNT(r.rental_id) AS rental_count
FROM film f
JOIN inventory i ON f.film_id = i.film_id
JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY f.film_id, f.title
ORDER BY rental_count DESC;
'''
df = pd.read_sql_query(num, engine)
df.head(10)
```

```python
num = '''WITH FilmTitles AS (
    SELECT film_id, title
    FROM film
),
RentalCounts AS (
    SELECT i.film_id, COUNT(*) AS rental_count
    FROM inventory i
    JOIN rental r ON i.inventory_id = r.inventory_id
    GROUP BY i.film_id
)
SELECT ft.film_id, ft.title, COALESCE(rc.rental_count, 0) AS rental_count
FROM FilmTitles ft
LEFT JOIN RentalCounts rc ON ft.film_id = rc.film_id
ORDER BY rental_count DESC;
'''
df = pd.read_sql_query(num,engine)
df.head(10)
```
