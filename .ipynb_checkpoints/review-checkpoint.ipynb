```python
import psycopg2
import yaml
from sqlalchemy import create_engine, inspect  # Import the inspect function

credential = '/home/ugochukwu/Desktop/multinational-retail-data-centralisation/db_creds.yaml'

class DatabaseConnector:
    def __init__(self, credential):
        self.credential = credential
        
    def read_db_creds(self):

        with open(self.credential, 'r') as yaml_file:
            credentials = yaml.safe_load(yaml_file)
        return credentials
    

    def init_db_engine(credentials):
        db_user = credentials['RDS_USER']
        db_password = credentials['RDS_PASSWORD']
        db_host = credentials['RDS_HOST']
        db_port = credentials['RDS_PORT']
        db_name = credentials['RDS_DATABASE']

        # Create a connection URL
        connection_url = f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'

        # Create and return the database engine
        engine = create_engine(connection_url)
        return engine
    
    def list_db_tables(engine):
        try:
            # Create a SQLAlchemy inspector to inspect the database
            inspector = inspect(engine)

            # Get a list of all table names in the database
            table_names = inspector.get_table_names()
            return table_names
        except Exception as e:
            print(f"Error listing database tables: {str(e)}")
            return None

# Initialize the database engine using init_db_engine
credentials = DatabaseConnector.read_db_creds(credential)
engine = DatabaseConnector.init_db_engine(credentials)

if engine is not None:
    # List all tables in the database
    tables = DatabaseConnector.list_db_tables(engine)

    if tables is not None:
        print("List of database tables:")
        for table in tables:
            print(table)
    else:
        print("Table listing failed.")

    # Don't forget to close the database connection when you're done
    engine.dispose()
else:
    print("Database engine initialization failed.")
```
